```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

def guess_number():
    target_number = random.randint(1, 9)

    # Create input box and button
    guess_box = widgets.BoundedIntText(
        value=1,
        min=1,
        max=9,
        description='Your Guess:'
    )
    button = widgets.Button(description="Submit Guess")
    output = widgets.Output()

    def on_button_click(b):
        with output:
            clear_output()
            user_guess = guess_box.value
            if user_guess == target_number:
                print(" Well guessed! The number was", target_number)
                button.disabled = True
                guess_box.disabled = True
            else:
                print(" Wrong guess. Try again!")

    button.on_click(on_button_click)
    display(guess_box, button, output)


guess_number()
```

```python
from IPython.display import display
import pandas as pd


results = []

for number in range(1, 51):
    if number % 3 == 0 and number % 5 == 0:
        results.append("FizzBuzz")
    elif number % 3 == 0:
        results.append("Fizz")
    elif number % 5 == 0:
        results.append("Buzz")
    else:
        results.append(str(number))


df = pd.DataFrame({'Number': range(1, 51), 'Result': results})
display(df)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

def binary_divisible_by_6():

    text = widgets.Text(
        value='0101,1000,1001,1010',
        description='Binaries:',
        placeholder='Enter comma-separated 4-digit binary numbers'
    )
    button = widgets.Button(description="Check Numbers")
    output = widgets.Output()

    def on_button_click(b):
        with output:
            clear_output()

            nums = [i.strip() for i in text.value.split(',')]


            res = [i for i in nums if int(i, 2) % 6 == 0]

            if res:
                print(" Binary numbers divisible by 6:", res)
            else:
                print(" No binary numbers divisible by 6 found.")

    button.on_click(on_button_click)
    display(text, button, output)


binary_divisible_by_6()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

def count_even_odd(numbers):
    even_count = 0
    odd_count = 0
    for num in numbers:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
    return even_count, odd_count

def even_odd_widget():
    # Create text box and button
    text = widgets.Text(
        value='1 2 3 4 5 6',
        description='Numbers:',
        placeholder='Enter numbers separated by spaces'
    )
    button = widgets.Button(description="Count")
    output = widgets.Output()

    def on_button_click(b):
        with output:
            clear_output()
            try:
                # Convert input into list of integers
                numbers = list(map(int, text.value.split()))
                even_count, odd_count = count_even_odd(numbers)
                print("✅ Number of even numbers:", even_count)
                print("✅ Number of odd numbers:", odd_count)
            except ValueError:
                print("⚠️ Please enter valid integers separated by spaces.")

    button.on_click(on_button_click)
    display(text, button, output)


even_odd_widget()
```
